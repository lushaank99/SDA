```python
import numpy as np
import random
from matplotlib import pyplot as plt 
import math
```

```python
def normal(x):
    a = (1/math.sqrt(2*math.pi))*math.exp((-1*x*x)/2)
    return a

def chi_squared(k, x):
    a = (1/(math.pow(2, k/2)*math.gamma(k/2)))*math.pow(x, (k-2)/2)*math.exp(-1*x/2)
    return a
```

```python
def chi_sq(n, c):
    st = []
    for i in range(n):
        d = []
        for j in range(c):
            d.append(np.random.normal(0,1))
        st.append(d)      
    np.square(st)
    a = np.sum(st, axis=0)
    return a

def normal_dis(c):
    d = []
    for j in range(c):
        d.append(np.random.normal(0,1))
    return d

def t_func(n, c):
    d = []
    for i in range(c):
        t = normal_dis(1)[0]
        ch = chi_sq(n, 1)
        t = t/math.sqrt(abs(ch)/n)
        d.append(t)
    return d
```

```python
da = normal_dis(1000)
da.sort()
plt.xlim(-5, 5)
plt.hist(da, bins=100, color="green")
plt.title("Z data")
plt.show()
deg = [3, 5, 10, 15]
for i in range(len(deg)):
    p = t_func(deg[i], 1000)
    p.sort()
    plt.xlim(-20, 20)
    plt.hist(p, bins=400)
    plt.title("T graph with " + str(deg[i]) + " degrees of freedom")
    plt.show()
    
```
