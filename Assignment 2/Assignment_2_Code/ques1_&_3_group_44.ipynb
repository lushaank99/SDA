```
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
from itertools import combinations
import seaborn as sns
```

```
no_of_sts = 5
```

```
sts = np.random.standard_normal((no_of_sts,1000))
```

```
sts
```

```
def get_sumOfSquares_and_x2(sts,n,single=False,plot=True):
  # st = np.random.standard_normal((n,1000))
  sum_squared_list=[]
  if single:
    combinations_list=[list(range(n))]
  else:  
    combinations_list=list(combinations(range(len(sts)), n))
  for combination in combinations_list:
    for index in combination:

      if index==0:
        sum_squared = np.square(sts[index])
      else:
        sum_squared = np.add(sum_squared,np.square(sts[index]))
    sum_squared_list.append(sum_squared)
    sum_squared=0      
  x2_samples = np.random.chisquare(n,1000)

  if not plot:
    return sum_squared_list[0] , x2_samples
  else:
    for s in sum_squared_list: 
      sns.distplot(s, hist=True)
      sns.distplot(x2_samples, hist=False)
      plt.legend(['Sum of squares','Chi-squared'])
      plt.show()     
```

```
get_sumOfSquares_and_x2(sts,2)
```

```
get_sumOfSquares_and_x2(sts,4)
```

```
def F(sts,n1,n2):
  u1,x1 = get_sumOfSquares_and_x2(sts,n1,single=True,plot=False)
  
  u2,x2 = get_sumOfSquares_and_x2(sts,n2,single=True,plot=False)
  
  num=u1/n1
  den = u2/n2
  c = num/den
  

  # return ((x1/n1)/(x2/n2))
  return c

```
