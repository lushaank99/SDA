# Exponential Distribution

```python
def exp_dis(x,lam):
  # print(x)
  if x<0:
    return 0
  else:
    return lam*(math.exp(-lam*x))
```

```python
#exponential distribution
lams = [0.5,1,2,5,10,15]
m=[]
v=[] 
x_values=[]
y_values=[]

for lam in lams:

  x = np.linspace(0,10,10000)
  y=[]

  for i in x:
    y.append(exp_dis(i,lam))
  y=np.array(y)

  mean=np.sum(np.multiply(x,y))*10/len(x)
  var = (np.sum(np.multiply(np.square(x),y))*10/len(x))-(mean**2)

  m.append(mean)
  v.append(var)
  x_values.append(x)
  y_values.append(y)
  
```

```python
x = np.linspace(0,6,10000)
y=[]

for i in x:
  y.append(exp_dis(i,2))
y=np.array(y)

# plt.figure(figsize=(20,10))

plt.plot(x,y)
plt.legend(['Lamda = 2'])
title_obj = plt.title('Exponential Distribution for Lambda = 2')            
plt.setp(title_obj, color='w',size=16.0)
plt.tick_params(axis='x', colors='white')
plt.tick_params(axis='y', colors='white')
plt.grid() 
plt.show()
```

```python
plt.figure(figsize=(20,10))
for x1,y1 in zip(x_values,y_values):
  plt.plot(x1,y1)
plt.legend(lams)
plt.tick_params(axis='x', colors='white')
plt.tick_params(axis='y', colors='white')
plt.grid() 
plt.show()
  
```

```python
data_exp = {'Lambda':lams,'Mean':m,'Varirance':v}
df_exp=pd.DataFrame(data_exp)
display(df_exp)
```

# Normal Distribution

```python
def normal_dis(mu,sigma,x):
  return (1 / (np.sqrt(2 * np.pi * np.power(sigma, 2)))) * (np.power(np.e, -(np.power((x - mu), 2) / (2 * np.power(sigma, 2)))))
```

```python
del x,m,v,x_values,y_values
del y
```

```python
mus = [-1,0,1]
sigmas =[0.5,1,2]
m=[]
v=[] 
const_mu=0
const_sigma=1
const_mu_x=[]
const_mu_y=[]
const_sigma_x=[]
const_sigma_y = []
x_values=[]
y_values=[]
legend=[]
mean_2=[]
var_2=[]
for mu in mus:
  for sigma in sigmas:
    legend.append('mu = '+str(mu)+" , sigma = "+str(sigma))
    x = np.linspace(mu-(sigma+4),mu+(sigma+4),1000000)
    y=normal_dis(mu,sigma,x)

    mean_prev=np.sum(np.multiply(x,y))
    mean = (mean_prev*10)/len(x)
    var_prev = np.sum(np.multiply(np.square(x),y))
    var = ((var_prev*10)/len(x))-(mean**2)


    m.append(round(mean,5))
    v.append(round(var,5))
    x_values.append(x)
    y_values.append(y)


    if sigma == const_sigma:
      const_sigma_x.append(x)
      const_sigma_y.append(y)
    if mu == const_mu:
      const_mu_x.append(x)
      const_mu_y.append(y)  
```

```python
plt.figure(figsize=(20,10))
for x1,y1 in zip(x_values,y_values):
  plt.plot(x1,y1)
plt.legend(legend)
plt.tick_params(axis='x', colors='white')
plt.tick_params(axis='y', colors='white')
plt.grid() 
plt.show()
```

```python
# plt.figure(figsize=(20,10))
for x1,y1 in zip(const_sigma_x,const_sigma_y):
  plt.plot(x1,y1)
plt.legend(mus)
title_obj = plt.title('Graph showing constant sigma and variable mu')            
plt.setp(title_obj, color='w')
plt.tick_params(axis='x', colors='white')
plt.tick_params(axis='y', colors='white')
plt.grid()
plt.show()
```

# Beta Distribution

```python
def beta_func(a,b):
  return (math.gamma(a)*math.gamma(b))/math.gamma(a+b)
  
```

```python
def beta_dist(a,b,x):
  return (1/beta_func(a,b))*(x**(a-1))*((1-x)**(b-1))
```

```python
del x,m,v,x_values,y_values,legend
del y
```

```python
alphas = [0.5,1,4]
betas =[0.5,1,1.5]
m=[]
v=[] 
const_alpha=1
const_beta=1.5
const_alpha_x=[]
const_alpha_y=[]
const_beta_x=[]
const_beta_y = []
x_values=[]
y_values=[]
legend=[]
for a in alphas:
  for b in betas:
    legend.append('alpha = '+str(a)+" , beta = "+str(b))
    x = np.linspace(0.1,0.9,100000)
    y=[]
    for i in x:
      y.append(beta_dist(a,b,i))
    y=np.array(y)  

    mean_prev=np.sum(np.multiply(x,y))
    mean = (mean_prev)/len(x)

    var_prev = np.sum(np.multiply(np.multiply(x,x),y))
    var = ((var_prev)/len(x))-(mean**2)


    m.append(round(mean,5))
    v.append(round(var,5))
    x_values.append(x)
    y_values.append(y)


    if b == const_beta:
      const_beta_x.append(x)
      const_beta_y.append(y)
    if a == const_alpha:
      const_alpha_x.append(x)
      const_alpha_y.append(y)  
```

```python
# plt.figure(figsize=(20,10))
for x1,y1 in zip(x_values,y_values):
  plt.plot(x1,y1)
plt.legend(legend)
plt.tick_params(axis='x', colors='white')
plt.tick_params(axis='y', colors='white')
plt.grid() 
plt.show()
```

```python
# plt.figure(figsize=(20,10))
for x1,y1 in zip(const_beta_x,const_beta_y):
  plt.plot(x1,y1)
plt.legend(alphas)
title_obj = plt.title('Graph showing constant beta and variable alpha')            
plt.setp(title_obj, color='w',size=12.0)
plt.tick_params(axis='x', colors='white')
plt.tick_params(axis='y', colors='white')
plt.grid()
plt.show()
```

```python
# plt.figure(figsize=(20,10))
for x1,y1 in zip(const_alpha_x,const_alpha_y):
  plt.plot(x1,y1)
plt.legend(betas)
title_obj = plt.title('Graph showing constant alpha and variable beta')            
plt.setp(title_obj, color='w',size=12.0)
plt.tick_params(axis='x', colors='white')
plt.tick_params(axis='y', colors='white')
plt.grid()
plt.show()
```

# Gamma Distribution

```python
def gamma_dist(a,b,x):
  return ((b**a)*(x**(a-1))*(math.exp(-1*b*x)))/math.gamma(a)
```

```python
def gamma2_dist(a,b,x):
  return ((b**a)*(x**(a-1))*(math.exp(-1*b*x)))/math.gamma(a)
```

```python
del x,y,m,v,x_values,y_values,legend,alphas,betas
```

```python
alphas = [0.5,1,1.5]
betas =[0.5,1,1.5]
m=[]
v=[] 
const_alpha=1
const_beta=1.5
const_alpha_x=[]
const_alpha_y=[]
const_beta_x=[]
const_beta_y = []
x_values=[]
y_values=[]
legend=[]
for a in alphas:
  for b in betas:
    legend.append('alpha = '+str(a)+" , beta = "+str(b))
    x = np.linspace(0.1,5,100000)
    y=[]
    for i in x:
      y.append(gamma_dist(a,b,i))
    y=np.array(y)  

    mean = a/b
    var = a/(b**2)


    m.append(round(mean,5))
    v.append(round(var,5))
    x_values.append(x)
    y_values.append(y)


    if b == const_beta:
      const_beta_x.append(x)
      const_beta_y.append(y)
    if a == const_alpha:
      const_alpha_x.append(x)
      const_alpha_y.append(y)  
```

# Lognormal Distribution

```python
def lognormal_dis(mu,sigma,x):
  return (np.reciprocal(x)) * (1 / (np.sqrt(2 * np.pi * np.power(sigma, 2)))) * (np.power(np.e, -(np.power((np.log(x) - mu), 2) / (2 * np.power(sigma, 2)))))
```

```python
del x,m,v,x_values,y_values
del y
```

```python
mus = [-1,0,1]
sigmas =[0.5,1,2]
m=[]
v=[] 
const_mu=0
const_sigma=1
const_mu_x=[]
const_mu_y=[]
const_sigma_x=[]
const_sigma_y = []
x_values=[]
y_values=[]
legend=[]
mean_2=[]
var_2=[]
for mu in mus:
  for sigma in sigmas:
    legend.append('mu = '+str(mu)+" , sigma = "+str(sigma))
    x = np.linspace(0.1,mu+(sigma+4),1000000)
    y=lognormal_dis(mu,sigma,x)


    mean_prev=np.sum(np.multiply(x,y))
    
    mean = math.exp(mu + ((sigma**2)/2))
    var = (math.exp(sigma**2)-1)*math.exp((2*mu)+(sigma**2))
  
    m.append(round(mean,5))
    v.append(round(var,5))
    x_values.append(x)
    y_values.append(y)


    if sigma == const_sigma:
      const_sigma_x.append(x)
      const_sigma_y.append(y)
    if mu == const_mu:
      const_mu_x.append(x)
      const_mu_y.append(y)  
```

```python
plt.figure(figsize=(20,10))
for x1,y1 in zip(x_values,y_values):
  plt.plot(x1,y1)
plt.legend(legend)
plt.tick_params(axis='x', colors='white')
plt.tick_params(axis='y', colors='white')
plt.grid() 
plt.show()
```
