```python
import numpy as np
import matplotlib.pyplot as plt 
import math
import scipy.stats as stats
```

# Exponential (Continuous Distribution)

```python
obser = 40
lam = 0.2
sim = 1000
```

```python
exp_Data = []
for i in range(sim):
    for j in range(obser):
        exp_Data.append(float(np.random.exponential(scale=1/lam, size=1)))
    
mean_exp_Data = []
for i in range(sim):
    mean_exp_Data.append(sum(np.random.exponential(scale=1/lam, size=obser))/obser)
    
```

```python
ran = (2, 8)
bins = 100
plt.hist(mean_exp_Data, bins, ran, color = 'blue', histtype = 'bar') 
plt.xlabel("value") 
plt.ylabel("count")
plt.title("Mean Calculation using CLT")
plt.show()
```

```python
t_mean = 1/lam
print("Exponential Data Mean is : ", round(t_mean, 2))
t_var = 1.0/(obser*lam*lam) 
print("Exponential Data Variance is : ", round(t_var, 4))

print("\n\n")

m = sum(mean_exp_Data)/sim
print("Experimental Normal Data Mean is : ", round(m, 2))
var = 0
for i in range(sim):
    var += pow(mean_exp_Data[i] - m, 2)
var = var/sim
print("Experimental Normal DataVariance is : ", round(var, 4))

s = math.sqrt(var)
l1 = m - s
l2 = m - 2*s
r1 = m + s
r2 = m + 2*s
```

```python
ran = (0, 20)
bins = 100
plt.hist(exp_Data, bins, ran, color = 'yellow', histtype = 'bar') 
plt.xlabel("value") 
plt.ylabel("count")
plt.title("Exponential Graph")
plt.show()

ran = (2, 8)
bins = 100
plt.hist(mean_exp_Data, bins, ran, color = 'green', histtype = 'bar') 
plt.axvline(x=m, ymin=0, ymax=1, color="r", linestyle="--")
plt.axvline(x=l1, ymin=0, ymax=1, color="r", linestyle="--")
plt.axvline(x=l2, ymin=0, ymax=1, color="r", linestyle="--")
plt.axvline(x=r1, ymin=0, ymax=1, color="r", linestyle="--")
plt.axvline(x=r2, ymin=0, ymax=1, color="r", linestyle="--")
plt.xlabel("value") 
plt.ylabel("count")
plt.title("Graph that is normal")

plt.show()
```
